```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
from datetime import datetime
```

```python
df_principal = pd.read_excel("/content/acoes_pura.xlsx", sheet_name = "Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/acoes_pura.xlsx", sheet_name = "Total_de_acoes")
df_principal
```

```python
df_ticker = pd.read_excel("/content/acoes_pura.xlsx", sheet_name = "Ticker")
```

```python
df_chat_gpt = pd.read_excel("/content/acoes_pura.xlsx", sheet_name = "ChatGPT")
```

```python
df_principal = df_principal[['Ativo',	'Data', 'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)': 'valor_final', 'Var. Dia (%)': 'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['var_dia_pct'] = df_principal['var_dia_pct'].apply(lambda x: "R${:,.2f}".format(x))
df_principal
```

```python
df_principal['valor_inicial'] = df_principal ['valor_final'] / (1 + df_principal['var_pct'])
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on = 'Ativo', right_on= 'Código', how='left' )
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código']).copy()
df_principal
```

```python
df_principal['variacao'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica': 'qtd_teorica'}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao'].apply(lambda x: 'Subiu' if x > 0 else ('Estavel' if x == 0 else 'Desceu'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on = 'Ativo', right_on= 'Ticker', how='left' )
df_principal = df_principal.drop(columns=['Ticker']).copy()
df_principal
```

```python
df_principal = df_principal[['Ativo', 'Nome', 'Data', 'valor_final', 'var_dia_pct', 'var_pct', 'valor_inicial', 'qtd_teorica', 'variacao', 'resultado']]
df_principal
```

```python
df_principal = df_principal.merge(df_chat_gpt , left_on = 'Nome', right_on= 'Nome da Empresa', how='left' )
df_principal = df_principal.drop(columns=['Nome da Empresa']).copy()
df_principal = df_principal.rename(columns={'Idade (em anos)': 'idade_anos'}).copy()
df_principal['qtd_teorica'] = df_principal['qtd_teorica'].apply(lambda x: format(x, ',d'))

df_principal
```

```python
'''
Tenho uma planilha com as colunas variacao e resultado. a partir dessas colunas monto a
tabela abaixo com uso de formulas do excel. Se eu tivesse a mesma tabela em um dataframe
do pandas como faria para chegar nos mesmos resultados? Na sua resposta, substitua df por df_principal.

Maior	R$ 4.762.926.995,25
Menor	-R$ 1.807.432.634,46
Media	R$ 165.190.210,48
Media de quem subiu	R$ 448.164.250,23
Media de quem desceu	-R$ 181.109.141,75
'''

# Vamos supor que você já tenha um DataFrame chamado df_principal com as colunas 'variacao' e 'resultado'

# Calculando a maior variação
maior_variacao = df_principal['variacao'].max()

# Calculando a menor variação
menor_variacao = df_principal['variacao'].min()

# Calculando a média da variação
media_variacao = df_principal['variacao'].mean()

# Calculando a média das variações positivas
media_variacao_positiva = df_principal[df_principal['resultado'] == 'Subiu']['variacao'].mean()

# Calculando a média das variações negativas
media_variacao_negativa = df_principal[df_principal['resultado'] == 'Desceu']['variacao'].mean()

# Exibindo os resultados
print("Maior\t", f"R$ {maior_variacao}")
print("Menor\t", f"R$ {menor_variacao}")
print("Media\t", f"R$ {media_variacao}")
print("Media de quem subiu\t", f"R$ {media_variacao_positiva}")
print("Media de quem desceu\t", f"R$ {media_variacao_negativa}")

```
